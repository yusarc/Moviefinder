```python
#Crosstab with pivot table
titels =  movies[['movieId', 'title']]

users_items = pd.pivot_table(data=ratings, 
                                 values='rating', 
                                 index='userId', 
                                 columns='movieId')
users_items.head()
```

```python
#Replace NaNs with zero
users_items.fillna(0, inplace=True)
users_items.head()
```

```python
#Compute cosine similarities
from sklearn.metrics.pairwise import cosine_similarity

user_similarities = pd.DataFrame(cosine_similarity(users_items),
                                 columns=users_items.index, 
                                 index=users_items.index)
user_similarities.head()
```

```python
# compute the weights for one user
user_id = 1

weights = (
    user_similarities.query("userId!=@user_id")[user_id] / sum(user_similarities.query("userId!=@user_id")[user_id])
          )
weights.head(6)
```

```python
# select movies that the unrated
not_rate_mov = users_items.loc[users_items.index!=user_id, users_items.loc[user_id,:]==0]
not_rate_mov.T
```

```python
# dot product between the unrated and the weights
weighted_averages = pd.DataFrame(not_rate_mov.T.dot(weights), columns=["predicted_rating"])
weighted_averages.head()
```

```python
#Merge with titles and show top 5
recommendations = weighted_averages.merge(titels, left_index=True, right_on='movieId').sort_values("predicted_rating", ascending=False).head()
recommendations
```

```python
def top_n_res(user_id,n):
    
    #Crosstab with pivot table
    titels =  movies[['movieId', 'title']]

    users_items = pd.pivot_table(data=ratings, 
                                 values='rating', 
                                 index='userId', 
                                 columns='movieId')
    #Replace NaNs with zero
    users_items.fillna(0, inplace=True)
   
    #Compute cosine similarities
    from sklearn.metrics.pairwise import cosine_similarity

    user_similarities = pd.DataFrame(cosine_similarity(users_items),
                                 columns=users_items.index, 
                                 index=users_items.index)
   
    # compute the weights for random user
    user_id=user_id
    weights = (
    user_similarities.query("userId!=@user_id")[user_id] / sum(user_similarities.query("userId!=@user_id")[user_id])
          )

    # select movies that the unrated
    not_rate_mov = users_items.loc[users_items.index!=user_id, users_items.loc[user_id,:]==0]
    not_rate_mov.T
   
   # dot product between the unrated and the weights
    weighted_averages = pd.DataFrame(not_rate_mov.T.dot(weights), columns=["predicted_rating"])
    
    #Merge with titles and show top n
    recommendations = weighted_averages.merge(titels, left_index=True, right_on='movieId').sort_values("predicted_rating", ascending=False).head(n)
    
    return list(recommendations["title"])
```

```python
top_n_res(327,5)
```
