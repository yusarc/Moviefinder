```python
ratings.head()
ratings= ratings[['movieId','rating']]
```

```python
ratings['rating_count']= ratings.groupby('movieId')['rating'].count()
ratings
```

```python
ratings[ratings['movieId']==1]
```

```python
ratings['score']= ratings['rating']*ratings['rating_count']
ratings
```

```python
ratings[(ratings['rating'] >=4) & (ratings['rating_count'] >=100)].sort_values("score", ascending=False).head(10)
```
